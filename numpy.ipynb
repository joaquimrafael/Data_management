````python
a = np.array([1,2,3])
b = np.array([1,-1,1])

c = np.array([[1,2,3], [0, 0, 0]])
d = np.array([[1,-1,1], [1, 2, 3]])

np.concatenate((a,b)), np.concatenate((c,d)), np.concatenate((c,d), axis=1), np.append(a,[4])
````

# Exercício 2

Faça um script para gerar a soma (100 somas) de 100 séries como acima (de 100 valores randômicos). Qual a distribuição dessas médias de valores?

> **O Teorema Central do Limite (TCL)** A soma de $N$ variáveis aleatórias independentes $X$, com qualquer distribuição e variâncias semelhantes, é uma variável com distribuição que se aproxima da distribuição normal quando $N$ aumenta.




````python
# seu código

tcl = np.array([])

for i in range(100):
  y = np.sum( np.random.rand(100))
  tcl = np.append(tcl, [y])
````

````python
sns.kdeplot(x=tcl, color='gray')
plt.title("TCL para 100 somas de 100 valores aleatórios")
plt.xlabel("Somas")
plt.ylabel("Ocorrência")
plt.show()
````
