Produza um gráfico de linha para uma função matemática de sua escolha.

1. Inclua um título com a função
2. Altere a cor e o estilo de linha do gráfico
3. Altere o tamanho da figura
4. Inclua rótulos para os eixos do gráfico

```python

def f(x):
  return x**2 - x*2

fig = plt.figure(figsize=(8,8))
x = np.arange(-10, 10, 1)
y = f(x)

print('x = ', x)
print('y = ', y)


plt.plot(x, y, color = 'g', linestyle='dashed', marker='o',markersize=6)


plt.title('Função f(x)= X ** 2 - X * 2', fontsize = 14, weight='bold')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend('y')


plt.show()
```
