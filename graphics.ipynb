# Basic imports

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
````

# **Relações dos Dados**

# Quick start

````python
df = sns.load_dataset("iris")
df.head()
````

````python
plt.plot(df['petal_width'], df['petal_length'])
plt.show()

plt.scatter(df['petal_width'], df['petal_length'], sizes=(20,250),s= df['petal_length'], alpha=0.6)
plt.show()
````

````python
sns.lineplot(x=df['petal_width'],y=df['petal_length'])
plt.show()

sns.scatterplot(x='petal_width', y='petal_length', data=df, hue='species')
plt.show()
````

# Exercícios

````python
df = pd.read_csv('https://raw.githubusercontent.com/Rogerio-mack/Visualizacao-de-Dados-em-Python/main/data/gapminder_2015.csv')
#df['cod_continent'] = df[['continent']].astype('category').apply(lambda x: x.cat.codes)
df.head()
````

# Exercício 1 (`plt.scatter`)

Faça uma figura com 3 gráficos envolvendo as variáveis
```
df['income_per_person'], df['LifeExpect']
```

1. Um plot de linha com o `Matplotlib`
2. Um scatter plot do `Matplotlib`
3. Um scatter plot do `Matplotlib` incluindo a dimensão dos dados `df['HappyIdx']` no tamanho dos pontos do gráfico de dispersão

Você pode empregar o modo `script layer` ou `artist layer`, como achar melhor.


````python
plt.figure(figsize=(17,8))

plt.suptitle('Exercício 1', fontsize = 14, weight ='bold', y=1.05)

plt.subplot(1,3,1)
plt.plot(df['income_per_person'],df['LifeExpect'], color='g')
plt.title("Gráfico de Linha")
plt.xlabel('Renda por pessoa', fontsize=14)
plt.ylabel('Expectativa de Vida',fontsize=10)

plt.subplot(1,3,2)
plt.scatter(df['income_per_person'], df['LifeExpect'], color ='r')
plt.title("Diagrama de dispersão")
plt.xlabel('Renda por pessoa', fontsize=14)
plt.ylabel('Expectativa de Vida',fontsize=10)

plt.subplot(1,3,3)
plt.scatter(df['income_per_person'], df['LifeExpect'], color='y', sizes=(15,260),s= df['HappyIdx'], alpha=0.5)
plt.title("Diagrama de dispersão (Tamanho das bolas: taxa de felicidade)")
plt.xlabel('Renda por pessoa', fontsize=14)
plt.ylabel('Expectativa de Vida',fontsize=10)




plt.tight_layout()
plt.show()
````

# Exercício 2 (`sns.lineplot`, `sns.scatterplot`)

Faça uma figura com 2 gráficos envolvendo as variáveis
```
df['income_per_person'], df['LifeExpect']
```

1. Um plot de linha com o `Seaborn`
2. Um scatter plot do `Seaborn`

Você pode empregar o modo `script layer` ou `artist layer`, como achar melhor.


````python
plt.figure(figsize=(18,6))

plt.suptitle('Exercício 2', fontsize = 14, weight ='bold', y=1.05)

plt.subplot(1,2,1)
sns.lineplot(x=df['income_per_person'],y= df['LifeExpect'])

plt.subplot(1,2,2)
sns.scatterplot(x='income_per_person', y='LifeExpect', data=df)

plt.tight_layout()
plt.show()
````

# Exercício 3 (`sns.scatterplot`)

Altere o segundo gráfico do exercício anterior para produzir um gráfico de dispersão incluindo a dimensão `df['continent']` dos dados (empregue o `Seaborn`).

````python
plt.figure(figsize=(18,6))

plt.suptitle('Exercício 3', fontsize = 14, weight ='bold', y=1.05)

plt.subplot(1,2,1)
sns.scatterplot(x='income_per_person',y='LifeExpect', data = df, hue='continent')

plt.subplot(1,2,2)
sns.lineplot(x='income_per_person',y='LifeExpect', data = df, hue='continent')

plt.show()
````
